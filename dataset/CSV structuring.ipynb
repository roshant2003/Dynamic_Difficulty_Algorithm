```python
import numpy as np

# Adding correct and total columns
df['total'] = np.random.randint(1, 101, size=len(df))
df['correct'] = np.random.randint(0, df['total'])

# Calculating difficulty_accuracy
df['diff_score'] = (100 - (df['correct'] / df['total']) * 100).round(2)

df.head()
```
